```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df1 is defined and contains the "tpep_pickup_datetime" column
# Convert the column to datetime
df1["tpep_pickup_datetime"] = pd.to_datetime(df1["tpep_pickup_datetime"])

# Calculate hour-wise counts
sd_h_dis = df1["tpep_pickup_datetime"].dt.day.value_counts()
sd_h_dis = sd_h_dis.sort_index()
sd_h_mean = sd_h_dis.mean()

# Function to add labels to the top of each bar
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2., 1.03 * height, '%s' % int(height))

# Create bar chart
rects = plt.bar(sd_h_dis.index, sd_h_dis.values)

# Plot mean line
plt.plot([0, len(sd_h_dis.index) + 1], [sd_h_mean, sd_h_mean], "r--")

# Set plot title and labels
plt.title("Pickup Datetime Daily Distribution")
plt.xlabel("Day")
plt.ylabel("Trips")

# Add grid
plt.grid()

# Add labels to the top of each bar
autolabel(rects)

# Display the plot
plt.show()
```
